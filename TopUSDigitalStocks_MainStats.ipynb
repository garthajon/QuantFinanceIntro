```python
import yfinance as yf
import pandas as pd
from tabulate import tabulate
from datetime import datetime, timedelta

# Define 25 top digital/AI/social media-related stocks
tickers = [
    "AAPL", "MSFT", "GOOGL", "AMZN", "META", "NVDA", "TSLA", "CRM", "ADBE", "INTC",
    "AMD", "NFLX", "SNOW", "PLTR", "SHOP", "SQ", "PYPL", "UBER", "LYFT", "COIN",
    "NET", "DDOG", "ZS", "OKTA", "MDB", "ROKU", "DOCU", "TWLO", "FSLY", "AI",
    "SMCI", "ASML", "PANW", "CRWD", "ABNB", "BIDU", "TWTR", "AVGO", "TEAM", "NOW",
    "ORCL", "TCEHY", "SE", "RBLX", "WIX", "ZI", "DUOL", "ESTC", "BL", "AMPL"
]

# Time window
end_date = datetime.today()
start_date = end_date - timedelta(days=365)

results = []

for ticker in tickers:
    data = yf.download(ticker, start=start_date, end=end_date)

    if data.empty:
        continue

    close_prices = data['Close']
    current_price = close_prices.iloc[-1]
    high_52wk = close_prices.max()
    low_52wk = close_prices.min()
    avg_52wk = close_prices.mean()
    std_52wk = close_prices.std()

    near_high = current_price >= (0.60 * high_52wk)
    near_low = current_price <= (1.4 * low_52wk)
    print (near_high)
    print (near_low)


    if (1==1)  or (2==2):
        results.append({
            "Ticker": ticker,
            "Current Price": round(current_price, 2),
            "52W High": round(high_52wk, 2),
            "52W Low": round(low_52wk, 2),
            "52W Avg": round(avg_52wk, 2),
            "52W Std Dev": round(std_52wk, 2)
            #"Near": "High" if (near_high)  else "Low"
        })

# Display results
df = pd.DataFrame(results)
#df_sorted = df.sort_values(by="Near", ascending=False)

df
```
